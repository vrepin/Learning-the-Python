```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=range(0,10)
y=x
y2=[i**2 for i in x]
plt.figure(figsize=(5, 5))
plt.subplot(2, 1, 1)
plt.plot(x, y) # построение графика
plt.title('Зависимости: y = x, y2 = x^2') # заголовок
plt.ylabel('y', fontsize=14) # ось ординат
plt.grid(True) # включение отображение сетки
plt.subplot(2, 1, 2)
plt.plot(x, y2) # построение графика
plt.xlabel('x', fontsize=14) # ось абсцисс
plt.ylabel('y2', fontsize=14) # ось ординат
plt.grid(True) # включение отображение сетки
```

```python
fruits = ['apple', 'peach', 'orange', 'bannana', 'melon']
counts = [34, 25, 43, 31, 17]
plt.figure(figsize=(5, 5))
plt.title('ddd')
plt.bar(fruits, counts)
plt.title('Fruits!')
plt.xlabel('Fruit')
plt.ylabel('Count')
```

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]
plt.figure(figsize=(5,6))
plt.subplot(2,2,1)
plt.plot(x, y1, '-')
plt.subplot(2,2,2)
plt.plot(x, y2, '--')
plt.subplot(2,2,3)
plt.plot(x, y3, '-.')
plt.subplot(2,2,4)
plt.plot(x, y4, ':')
```

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]
fig, axs = plt.subplots(2,2,figsize=(7,7))
axs[0,0].set_xlabel('hhh')
axs[0,0].plot(x, y1, '-', label='y1')
axs[0,1].plot(x, y2, '--', label='y2')
axs[1,0].plot(x, y3, '-.', label='y3')
axs[1,1].plot(x, y4, ':', label='y4')
axs[0,0].legend()
```
